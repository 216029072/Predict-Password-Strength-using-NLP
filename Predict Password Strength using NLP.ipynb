```python
password_tuple=np.array(data)
```

```python
import random 
random.shuffle(password_tuple)
```

```python
x=[labels[0] for labels in password_tuple]
y=[labels[1] for labels in password_tuple]
```

```python
x
```

```python
y
```

```python
#funtion to put words in a list 

def word_divide_char(inputs):
    character=[]
    for i in inputs:
        character.append(i)
    return character 
```

```python
word_divide_char('prisonbreak1')
```

```python
from sklearn.feature_extraction.text  import TfidfVectorizer 
```

```python
vectorizer=TfidfVectorizer(tokenizer=word_divide_char)
```

```python
X=vectorizer.fit_transform(x)
```

```python
X.shape
```

```python
vectorizer.get_feature_names()
```

```python
X
```

```python
first_document_vector=X[0]
first_document_vector
```

```python
first_document_vector.T.todense()
```

```python
#Taking Matrics IDf Index and putting into a dataframe

df=pd.DataFrame(first_document_vector.T.todense(),index=vectorizer.get_feature_names(),columns=['TF-IDF'])
```

```python
#Sorting Values according to the importance 
df.sort_values(by=['TF-IDF'],ascending=False)

#Therefore we can see that '7'is a important icon in the document
```

```python
#Training the data
from sklearn.model_selection import train_test_split 
```

```python
#Spliting test data and using only 20% for testing , the rest for training 

X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2)
```

```python
#Checking training data

X_train.shape
```

```python
#importing logistic regression class
from sklearn.linear_model import LogisticRegression 
```

```python
clf=LogisticRegression(random_state=0,multi_class='multinomial')
```

```python
clf
```

```python
#fitting data
clf.fit(X_train,y_train)
```

```python
#creating a password and predicting how strong it is 

dt=np.array(['%@123abcd'])
pred=vectorizer.transform(dt)
clf.predict(pred)
```

```python
#making prediction using data 

y_pred=clf.predict(X_test)
y_pred
```

```python
#checking the accuracy of the model 

from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm=confusion_matrix(y_test,y_pred)
```

```python
#displaying false and true predictions 
cm
```

```python
#getting the accuracy 
accuracy=accuracy_score(y_test,y_pred)
```

```python
#display of accuracy
accuracy
```

```python

#Classification report 
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
